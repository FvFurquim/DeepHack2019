Despesas Tabela Teste:

```
dfTest['vl_despesa'] = (dfTest['vl_despesa'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
dfTest['vl_despesa'] = dfTest.apply(lambda x: x['vl_despesa'] / 100, axis = 'columns')

dfTest_event_paid = dfTest[dfTest.evento == 'Valor Pago']
dfTest_event_emp = dfTest[dfTest.evento == 'Empenhado']
dfTest_event_annul = dfTest[dfTest.evento == 'Anulação']

TotalEmp = dfTest_event_emp['vl_despesa'].sum()
TotalAnnul = dfTest_event_annul['vl_despesa'].sum()
TotalPaid = dfTest_event_paid['vl_despesa'].sum()

TotalEmpVal = TotalEmp - TotalAnnul

Debt = TotalEmpVal - TotalPaid

print('Total: ' + f'{TotalPaid:.2f}')

if(Debt <= 0):
  print("Divida: " + "Não " + f'{Debt:.2f}')
else:
  print("Divida: " + "Sim " + f'{Debt:.2f}')
```

Despesa anual Tabela Teste:

```
mes = 1
ano = 2018
totalAnual = 0.00
debtAnual = 0.00

while(mes <= 12):
  jsonAddressTest = 'https://transparencia.tce.sp.gov.br/api/json/despesas/' + str(munNameAda) + '/' + str(ano) + '/' + str(mes)
  dfTest = pd.read_json(jsonAddressTest)

  dfTest['vl_despesa'] = (dfTest['vl_despesa'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
  dfTest['vl_despesa'] = dfTest.apply(lambda x: x['vl_despesa'] / 100, axis = 'columns')

  dfTest_event_paid = dfTest[dfTest.evento == 'Valor Pago']
  dfTest_event_emp = dfTest[dfTest.evento == 'Empenhado']
  dfTest_event_annul = dfTest[dfTest.evento == 'Anulação']

  TotalEmp = dfTest_event_emp['vl_despesa'].sum()
  TotalAnnul = dfTest_event_annul['vl_despesa'].sum()
  TotalPaid = dfTest_event_paid['vl_despesa'].sum()

  TotalEmpVal = TotalEmp - TotalAnnul
  Debt = TotalEmpVal - TotalPaid
  
  totalAnual = totalAnual + TotalEmpVal
  debtAnual = debtAnual + Debt
  
  mes = mes + 1

print('Total: ' + f'{totalAnual:.2f}')

if(debtAnual <= 0):
  print("Divida: " + "Não " + f'{debtAnual:.2f}')
else:
  print("Divida: " + "Sim " + f'{debtAnual:.2f}')
```

```
jsonAddressAllMun = "https://transparencia.tce.sp.gov.br/api/json/municipios"
dfMun = pd.read_json(jsonAddressAllMun)
dfMun
```

```
posMun = 0
ano = 2018

dfCreation = {'municipio': [], 'despesas': [], 'divida': []}
dfDespAnual = pd.DataFrame(data = dfCreation)

while(posMun <= 643):
  munName = dfAllMun.loc[posMun].municipio
  mes = 1
  totalAnual = 0.00
  debtAnual = 0.00
  
  while(mes <= 12):
    jsonAddressMun = 'https://transparencia.tce.sp.gov.br/api/json/despesas/' + str(munName) + '/' + str(ano) + '/' + str(mes)
    dfMun = pd.read_json(jsonAddressMun)

    dfMun['vl_despesa'] = (dfMun['vl_despesa'].str.split()).apply(lambda x: float(x[0].replace(',', '')))
    dfMun['vl_despesa'] = dfMun.apply(lambda x: x['vl_despesa'] / 100, axis = 'columns')

    dfMun_event_paid = dfMun[dfMun.evento == 'Valor Pago']
    dfMun_event_emp = dfMun[dfMun.evento == 'Empenhado']
    dfMun_event_annul = dfMun[dfMun.evento == 'Anulação']

    TotalEmp = dfMun_event_emp['vl_despesa'].sum()
    TotalAnnul = dfMun_event_annul['vl_despesa'].sum()
    TotalPaid = dfMun_event_paid['vl_despesa'].sum()

    TotalEmpVal = TotalEmp - TotalAnnul
    Debt = TotalEmpVal - TotalPaid
    
    totalAnual = totalAnual + TotalEmpVal
    debtAnual = debtAnual + Debt
    
    mes = mes + 1
    
  dfDespAnual = dfDespAnual.append({'municipio' : munName, 'despesas': totalAnual, 'divida': debtAnual} , ignore_index=True)

  posMun = posMun + 1
  
dfDespAnual
```
